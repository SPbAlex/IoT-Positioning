```python
import pandas as pd
import numpy as np

from keras.layers import Input, Dense, Flatten, Reshape, Add, Activation,  Conv2D, BatchNormalization
from keras.layers import MaxPooling2D, UpSampling2D, LSTM, Dropout, ConvLSTM2D
from keras.models import Model
from keras import backend as K
```

```python
df = pd.read_csv('stat.csv',sep=';')
```

```python
df
```

```python
encoding_dim = 16
input_data = Input(shape=(2,10,))
l1 = Dense(encoding_dim, activation='sigmoid')(input_data)
l2 = Dense(encoding_dim/2, activation='sigmoid')(l1)
l3 = Dense(encoding_dim/4, activation='sigmoid')(l2)
flat = Flatten()(l3)
output_data = Dense(1, activation='sigmoid')(flat)

net = Model(input_data, output_data, name="network")
net.compile(optimizer='adam', loss='mse')
net.summary()
```

```python
x = df.drop('Distance',1)
y = df['Distance']
a = np.array(x)
b = np.array(y)
b = np.float_(b)
a = np.float_(a)

a_min = np.amin(a)
a_max = np.amax(a)
b = b / np.amax(b)
a = a / np.amax(a)
a = a[0:620]
b = b[0:620]
b_ = b[0:62]
for b_index in range(620,10):
    if b_index % 10 == 0:
        b_ = b[b_index]

a = a.reshape(62,2,10)
b = b_
```

```python
epoch = 500
batch = 30
net.fit(a, b,epochs=epoch,batch_size=batch,shuffle=True)
```
